## Detecção e remoção de outliers

```python
#Detecção de outliers utilizando o método KNN

from pyod.models.knn import KNN
detector = KNN()
detector.fit(X_train.iloc[:,3:13])

# get the prediction labels and outlier scores of the training data
previsoes = detector.labels_ # binary labels (0: inliers, 1: outliers)
confianca_previsoes = detector.decision_scores_ # raw outlier scores

outliers = []
for i in range(len(previsoes)):
    if previsoes[i] == 1: # Se na previsao for 1, valor é considerado outlier
        outliers.append(i)
        
lista_outliers = X_train.iloc[outliers, :]  # lista dos indices dos outliers
```

```python
outliers_series = lista_outliers.index // 128 # Indices para series 
outliers_series = np.unique(np.array(outliers_series)) # Remove series repetidas do vetor
```

```python
# Remove as series detectadas com valores outliers da base de dados
for i in range(len(outliers_series)):
    X_train = X_train.drop(X_train[ X_train['series_id'] == outliers_series[i]].index)
    y_train = y_train.drop(y_train[y_train['series_id'] == outliers_series[i]].index)

print('Quantidade de series detectadas com valores outliers: ', len(outliers_series))
```

## Pré-processamento dos previsores

```python
# Primeiro será removido o row_id, series_id e measurement_number que não são variavie previsores
X_train = X_train.iloc[:,3:13]

# Os dados são normalizados no intervalo [0,1]
from sklearn.preprocessing import MinMaxScaler
normalizador = MinMaxScaler(feature_range=(0,1))
X_train_norm = normalizador.fit_transform(X_train)

# Como cada série é composta por 128 leituras, as todas as leituras referentes a cada série foram colocadas em uma array
previsores = []
for i in range(0, y_train.shape[0]):
    previsores.append(X_train_norm[(i*128):((i+1)*128),0:10])

previsores = np.array(previsores)
```

## Pré-processamento das classes

```python
# Primeiro será removido o series_id e group_id deixando apenas a classe (surface)
classe = y_train.iloc[:,2].values

# Como as classes são strings e o classificador que será utilizado aceita apenas valores numéricos, 
#as strings foram transformadas para variáreis dummy

#Transforma cada classe para um número
#e.g: carpet - 0 
#     concrete - 1
#     fine_concrete - 2
#     ....
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
classe = labelencoder.fit_transform(classe)

print(labelencoder.classes_)
```

## Divisão em treinamento e teste

A base de treinamento será utilizada para encontrar o melhor modelo através de validação cruzada e estimar seu desempenho. Posteriormente será utilizada para treinar o m final.
A base de teste (que nunca foi vista pelo modelo) será utilizada para testar o modelo final.

```python
from sklearn.model_selection import train_test_split
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size=0.2, random_state=0)

import collections
print('Amostras treinamento:' ,classe_treinamento.shape[0])
counter = collections.Counter(classe_treinamento)
print(sorted(counter.items()))
print('Amostras teste:' ,classe_teste.shape[0])
counter = collections.Counter(classe_teste)
print(sorted(counter.items()))
```

A quantidade de amostras de cada classe está desbalanceada,  o que irá afetar a precisão do classificador. 
Para lidar com este problema, os dois métodos mais comuns de serem aplicados são: Oversampling e undersampling.

Oversampling: consiste em duplicar as observações da classe minoritária para obter um conjunto de dados balanceado, deixando a base de dados com amostras repetidas. Como a classe minoritária (hard_tiles) possui apenas 21 registros, a aplicação da técnica de oversampling replicaria os registros dessa classe várias vezes (aproximadamente 37 vezes), podendo causar overfitting.

Undersampling: Aleatoriamente remove amostras da classe majoritária, com ou sem substituição. Esta técnica pode aumentar a variância do classificador e pode potencialmente descartar amostras úteis ou importantes. Como a classe minoritária (hard_tiles) possui apenas 21 registros, a aplicação da técnica de undersampling deixaria a base de dados com poucas amostras de cada classe, dificultando o aprendizado. 

Outra abordagem é atribuir um peso para cada classe, de forma a atribuir um peso maior para a classe minoritária (e conseguentemente uma punição maior em caso de erro) em relação a classe majoritária. Nesta abordagem não é necessário o equilibrio da base de dados, pois a diferença de pesos atribuido a cada classe compensa o desbalanceamento. Esta última abordagem foi a adotada. 

```python
# Calculo dos pesos atribuidos a cada classe no conjunto de treinamento para compensar o desequilibrio da base de dados
from sklearn.utils import class_weight

class_weights = class_weight.compute_class_weight('balanced', np.unique(classe_treinamento), classe_treinamento)

print('Peso de cada classe:')
print(class_weights)
```

```python
#Transforma a classe de número para dummy
#e.g:  carpet - 0000001
#     concrete - 0000010
#     ....

from keras.utils import np_utils
classe_teste_dummy = np_utils.to_categorical(classe_teste)
classe_treinamento_dummy = np_utils.to_categorical(classe_treinamento)
```

# Treinamento do modelo de Machine Learning

## Definição do modelo

```python
from keras import backend as K

def sensitivity(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    return true_positives / (possible_positives + K.epsilon())

def specificity(y_true, y_pred):
    true_negatives = K.sum(K.round(K.clip((1-y_true) * (1-y_pred), 0, 1)))
    possible_negatives = K.sum(K.round(K.clip(1-y_true, 0, 1)))
    return true_negatives / (possible_negatives + K.epsilon())


from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv1D, MaxPooling1D, GlobalMaxPooling1D, LSTM, Bidirectional
from keras.layers.normalization import BatchNormalization

def criar_rede():
    classificador = Sequential()
    classificador.add(Conv1D(filters = 128, kernel_size = 8, activation = 'relu',
                             input_shape=(previsores.shape[1], previsores.shape[2])))
    classificador.add(BatchNormalization())
    classificador.add(MaxPooling1D(pool_size = 2))
    
    classificador.add(Conv1D(filters = 64, kernel_size = 8, activation = 'relu'))
    classificador.add(BatchNormalization())
    classificador.add(MaxPooling1D(pool_size = 2))
    
    classificador.add(LSTM(units = 80, return_sequences = True))
    classificador.add(Dropout(0.2))
        
    classificador.add(LSTM(units = 50, return_sequences = False))
    classificador.add(Dropout(0.2))
    
    classificador.add(Dense(units = 9, activation='softmax'))
    
    classificador.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                      metrics = ['categorical_accuracy', sensitivity, specificity])
    
    return classificador
```

```python
from keras.wrappers.scikit_learn import KerasClassifier

classificador = KerasClassifier(build_fn = criar_rede,
                                epochs = 300,
                                batch_size = 400)
```

```python
from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

es = EarlyStopping(monitor = 'loss', min_delta = 1e-3, patience = 10) # Monitora a funcao de loss e se em 10 epocas ela nao melhorar em pelo menos 1e-3 o treinamento termina
rlr = ReduceLROnPlateau(monitor = 'loss', factor = 0.2, patience = 5) # Diminui a taxa de aprendizagem se ficar 5 epocas sem melhorar o loss
mcp = ModelCheckpoint(filepath = 'pesos.h5', monitor = 'loss', save_best_only = True, verbose = 1) # Salva o melhor modelo que encontrar
```

### Avaliação cruzada para estimar o desempenho do classificador 

```python
from sklearn.model_selection import cross_val_score

resultados = cross_val_score(estimator = classificador,
                             X = previsores_treinamento, y = classe_treinamento,
                             cv = 10, scoring = 'accuracy', fit_params= {'callbacks': [es, rlr], 'class_weight' : class_weights, 'verbose': 2}, verbose = 1)
```

```python
media = resultados.mean()
desvio = resultados.std()

print('Media:', media)
print('Desvio padrão:', desvio)
```

### Treinamento modelo final
Para treinar o modelo final será utilizado o melhor modelo encontrado durante os testes com validação cruzada.

```python
classificador = criar_rede()
rnn = classificador.fit(previsores_treinamento, classe_treinamento_dummy, epochs = 300, batch_size = 400, validation_data = (previsores_teste, classe_teste_dummy) ,callbacks=[es, rlr, mcp], class_weight = class_weights, verbose = 2)
```

```python
print('Configuração do melhor modelo encontrado:')
classificador.summary()
```
